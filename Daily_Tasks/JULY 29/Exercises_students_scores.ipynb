```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Create and Read CSV
csv_data = """Name,Math,Science,English
Amit,78,85,74
priya,92,89,96
kiran,65,70,60
anita,88,95,90
Ravi,55,60,58
"""
```

```python
# Save the data as students.csv
with open('students.csv', 'w') as f:
  f.write(csv_data)
```

```python
# Read the CSV into a DataFrame
df = pd.read_csv('students.csv')
```

```python
# Print the first 3 rows
print("First 3 rows of the dataset:")
print(df.head(3))
```

```python
#Step 2: Calculate Average Scores
df['average'] = df[['Math', 'Science', 'English']].mean(axis=1)
top_student = df.loc[df['average'].idxmax()]
print("\nStudent with highest average score:")
print(top_student)
```

```python
# Step 3: Filter Students
print("\nStudents with Science score > 80: ")
print(df[df['Science']> 80])
print("\nStudents with average score <70:")
print(df[df['average']<70])
```

```python
# Step 4: Sorting
print("\nTop 3 students by English score (descending):")
print(df.sort_values(by ='English', ascending=False).head(3))
```

```python
# Step 5: Chart Exercise

plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['average'], color ='skyblue')
plt.xlabel('Students')
plt.ylabel('Average Score')
plt.title('Average Scores of students')
plt.tight_layout()
plt.show()
```

```python
df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o', figsize=(10,6))
plt.title("Subject_Wise Scores of Students")
plt.ylabel("Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Bonus Task
subject_totals = df[['Math', 'Science', 'English']].sum()
top_subject = subject_totals.idxmax()
print(f"\nSubjects with highest total score:top{top_subject}")
# Save updated DataFrame with Average column
df.to_csv('students_Summary.csv', index=False)
```
